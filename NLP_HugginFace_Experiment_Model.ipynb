#Zero shot Classification

This model  helps us to identify the nature of the work, which belongs to the category, so we can use the category with own created and imagine and distinguish between the nature of labels.

So it can be used in the NEWS forecasting and developing the concern for highlighting the news attributes belonging.

Below is the Some of the code and answers, which can help to distinguish and follow the concern for the role

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

zeroshot = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
zeroshot(sequence_to_classify, candidate_labels)
```

```python
candidate_labels = ['travelling', 'Buildings', 'Watching', 'driving']
zeroshot("I am travelling to Dubai", candidate_labels)
```

```python
text = "I think Prime minister MR. Narendra Modin who worked for the india as well as for all over the world, including the rising the GDP and connection between the foreign countries are good and stable while the MR. Jay shankar who is the relation forigen minister of india welcomed all President and the allies into the nature of work and peace "
candidate_labels = ['News', "Affairs", "Work", "Name"]
zeroshot(text, candidate_labels)
```

```python
text = "In Delhi 5 people found dead in one house at the same time, Self attack or suicide attack could be the reason behind it"
candidate_labels = ['News', "Murder", "Crime", "Politics"]
zeroshot(text, candidate_labels)
```
