```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification") # if we do not provide the model name, then it will be choosen as deafult moodel
```

```python
pipe("Hii my name is Abhijeet, I teach how to")
```

```python
#The output means that the model predicted the text "Hii my name is Abhijeet, I teach hoe to" as having a **positive sentiment** with a confidence score of **92.66%**. This score reflects how confident the model is
# about the prediction being positive.
```

In this above output:

- **Label: "POSITIVE"** means the model classified the text as having a positive sentiment or tone. This can be interpreted as the text giving off a generally positive feeling or context.
  
- **Score: 0.9266 (or 92.66%)** represents the model's confidence in its prediction. A higher score indicates the model is more confident that the text is positive. In this case, it's 92.66% sure that the tone of the text is positive.

Even though your statement might not seem overly positive, the model interprets the tone and context of the words used and still detects a positive sentiment.

#Zero shot Classification

```python
zeroshot = pipeline("zero-shot-classification")
```

SO the Zero short classification is:
Zero-shot classification is a machine learning technique that allows a model to classify data into multiple classes without being trained on specific examples for those classes. It's a subfield of zero-shot learning, which is a paradigm that allows models to generalize to new categories based on what they've learned during training.

For example, the Contrastive Language-Image Pretraining (CLIP) model was trained on over 400 million pairs of images and text. CLIP can classify images by common objects or by characteristics of an image. For example, if you pass an image of a Toyota Camry through CLIP with the prompts “Toyota” and “Mercedes”, CLIP will identify the model as a Toyota car.


Zero-shot classification differs from single or few-shot classification, which include a single or a few examples of the selected task.

```python
zeroshot("My stocks are going down just because of having international politics instability",
          candidate_labels =["Education", "Politics", "Business"])
```

```python
zeroshot("My stocks are going down just because of indin politics",
          candidate_labels =["Education", "Politics", "Business"])
```

```python
zeroshot("The indian budget is locating the most drastic postive change in the whole world as economic wise, so the GDP of india is changing vast",
          candidate_labels =["International Market", "Politics", "Business"])
```

```python
zeroshot("I am learning data science and i want my skiils to be used in orgnaisation so that i can make good money using data science",
          candidate_labels =["Goverment orgnaisation", "Companies", "Business"])
```
